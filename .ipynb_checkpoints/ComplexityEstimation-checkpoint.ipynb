```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import tensorflow as tf
import scipy
import matplotlib.pyplot as plt
import keras
from matplotlib.image import imread

import complexity_estimation_release as complexity

# Set up image folder and trained model weights file
# You need to download and decompress the PASCAL VOC dataset, then set its path here

pascal_voc_folder = 'D:/Dropbox/CogShift-light/VOCdevkit 3/VOC2012/JPEGImages/'
model_path = "trained_model_inception_v3.h5"
```

```python
# Load some test images frm PASCAL VOC
images, mus, keys, urls = complexity.load_demo_data(pascal_voc_folder)
```

```python
# Remove small images
images,mus,keys = complexity.remove_small_images(images,mus,keys)
```

```python
# Resize, convert to float32, multiply by 255, zero-centre and set to channel-first ordering
X_val = complexity.prep_x_resize_vgg(images,500)
```

```python
# Inspect the model
model.summary()
```

```python
# Predict complexity values for the demo images and plot them against their ground truth; 
# you should obtain Pearson's R at least 0.81.
complexity.visual_correlation(model, X_val, Y_val)
```

```python
# Process your own images - set up data

N = 5

filenames = [pascal_voc_folder + '2008_000912.jpg',
 pascal_voc_folder + '2008_002621.jpg',
 pascal_voc_folder + '2008_000060.jpg',
 pascal_voc_folder + '2008_000629.jpg',
 pascal_voc_folder + '2008_000003.jpg']

images = []

for f in filenames:
  images.append(imread(f))
```

```python
# Resize, convert to float32, multiply by 255, zero-centre and set to channel-first ordering

X = complexity.prep_x_resize_vgg(images,500)
```

```python
# Predict complexity values
Y = model.predict(X)
```

```python
# Show images and predicted values

for i in range(N):
  fig,ax = plt.subplots()
  ax.imshow(images[i])
  score = np.round(Y[i][0],2)
  ax.set_xlabel('Complexity = '+str(score))
```
